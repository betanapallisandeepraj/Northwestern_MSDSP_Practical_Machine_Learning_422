```python
import pandas as pd

train_data = pd.read_csv("titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("titanic/test.csv")
test_data.head()
```

```python
import seaborn as sns
sns_plot=sns.pairplot(train_data)
sns_plot.savefig("titanic_pairplot.png")
```

```python
col = train_data.columns[0:11]
col
df_corr = train_data[col]
corrmat = df_corr.corr()

import numpy as np
mask = np.triu(np.ones_like(corrmat, dtype=bool))

import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(12,12))
sns.heatmap(corrmat, vmax=1, square=True, mask=mask, annot=True, cmap='RdYlBu', linewidths=1)
plt.savefig("titanic_heatmap.png")
```

```python
sns_plot=sns.pairplot(train_data, hue='Survived')
sns_plot.savefig("titanic_pairplot_survived.png")
```
